```python
import requests
import hmac
import hashlib
import time

API_KEY = 'jsVd7H3RswsiGR6kU8mpaVpz6NWIpWQ0X6s7IVeNrg7DQ6zPUyWXEpH5gAKBfyeC'
API_SECRET = 'XzVg3q6BHC2qLXbGIrgLJvfDw6pC6FS42S0rzjvFdA91fVnAQfp7t9pAh5TTvVdc'

def get_binance_data(endpoint):
    base_url = 'https://fapi.binance.com'
    timestamp = int(time.time() * 1000)
    query_string = f'timestamp={timestamp}'
    signature = hmac.new(API_SECRET.encode(), query_string.encode(), hashlib.sha256).hexdigest()
    headers = {
        'X-MBX-APIKEY': API_KEY
    }
    url = f'{base_url}{endpoint}?{query_string}&signature={signature}'
    response = requests.get(url, headers=headers)
    return response.json()

# Example usage
account_info = get_binance_data('/fapi/v2/account')
trades = get_binance_data('/fapi/v1/userTrades')
income = get_binance_data('/fapi/v1/income')

print(account_info)
print(trades)
print(income)
```

```python
from binance.client import Client 
API_KEY = 'jsVd7H3RswsiGR6kU8mpaVpz6NWIpWQ0X6s7IVeNrg7DQ6zPUyWXEpH5gAKBfyeC' 

API_SECRET = 'XzVg3q6BHC2qLXbGIrgLJvfDw6pC6FS42S0rzjvFdA91fVnAQfp7t9pAh5TTvVdc' 

client = Client(API_KEY, API_SECRET)
```

```python
symbol = 'BTCUSDT'
end_time = datetime.now()
start_time = end_time - timedelta(days=90)
```

```python
klines = client.get_historical_klines(
    symbol=symbol,
    interval=Client.KLINE_INTERVAL_1DAY,  # Daily intervals
    start_str=start_time.strftime("%d %b, %Y"),
    end_str=end_time.strftime("%d %b, %Y")
)
```

```python
df = pd.DataFrame(klines, columns=[
    'Open Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'Close Time',
    'Quote Asset Volume', 'Number of Trades', 'Taker Buy Base Asset Volume',
    'Taker Buy Quote Asset Volume', 'Ignore'
])
```
